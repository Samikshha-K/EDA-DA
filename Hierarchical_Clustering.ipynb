# **Hierarchical Clustering**

Importing libraries and dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/salemprakash/EDA/main/Data/Benefits.csv')
```

```python
plt.figure(figsize=(8,5))
plt.title("Tenure",fontsize=16)
plt.xlabel ("Tenure(years)",fontsize=14)
plt.grid(True)
plt.hist(df['tenure'],color='orange',edgecolor='k')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Age",fontsize=16)
plt.xlabel ("Age(Years)",fontsize=14)
plt.grid(True)
plt.hist(df['age'],color='green',edgecolor='k')
plt.show()
```

```python
X = df.iloc[:,[4,5]].values
```

```python
import scipy.cluster.hierarchy as sch

plt.figure(figsize=(15,6))
plt.title('Dendrogram')
plt.ylabel('Euclidean distances')
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.show()
```
